```python
import seaborn as sb
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
df_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data_Sets/Country-data.csv')
df_data
```

```python
df = df_data.loc[:, df_data.columns != "country"]
```

```python
X = df.loc[:, df.columns != "gdpp"]
y = df["gdpp"]
```

```python
X
```

```python
kmn = KMeans(n_clusters=3)
kmn.fit(X)
labels = kmn.predict(X)
```

```python
labels
```

```python
df_prediction = pd.DataFrame(X, columns=X.columns)
df_prediction
```

```python
df_prediction['clusters1']=labels
df_prediction
```

```python
xs=df_prediction['inflation']
ys=df_prediction['income']
centroids = kmn.cluster_centers_
plt.scatter(xs, ys, c=labels)
plt.xlabel('inflation')
plt.ylabel('income')
```

```python
kmn.inertia_
```

```python
xs=df_prediction['exports']
ys=df_prediction['imports']
centroids = kmn.cluster_centers_
plt.scatter(xs, ys, c=labels)
plt.xlabel('exports')
plt.ylabel('imports')
```

```python
inertia_list = []
for k in np.arange(1, 19):
  kmn = KMeans(n_clusters = k)
  kmn.fit(X)
  inertia_list.append(kmn.inertia_)
```

```python
plt.plot(np.arange(1,19),inertia_list,'o-',c='red')
plt.xlabel('number of clusters')
plt.ylabel('Inertia')
plt.show()
```
